```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
grads = pd.read_csv('recent-grads.csv')
print(grads.iloc[0])
print()
print(grads.head())
print()
print(grads.tail)
```

#### OVERVIEW OF THE DATASET

**Rank** - Rank by median earnings (the dataset is ordered by this column).

**Major_code** - Major code.

**Major** - Major description.

**Major_category** - Category of major.

**Total** - Total number of people with major.

**Sample_size** - Sample size (unweighted) of full-time.

**Men** - Male graduates.

**Women** - Female graduates.

**ShareWomen** - Women as share of total.

**Employed** - Number employed.

**Median** - Median salary of full-time, year-round workers.

**Low_wage_jobs** - Number in low-wage service jobs.

**Full_time** - Number employed 35 hours or more.

**Part_time** - Number employed less than 35 hours

```python
print(grads)
```

```python
print(grads.shape)
print(grads.describe())
print()
print(grads.info())
```

- The dataset consist of 21 columns and 173 rows, some of the rows has null values.
- The null values are very little and therefore dropping them will not affect our analysis.

**DROPPING NULL VALUES**

```python
grads = grads.dropna()
print(grads.shape)
print()
print(grads.isnull().sum())
```

- The rows with the null values has been dropped and the dataset now has 21 cloumns and 172 rows.

#### GENERATING HISTOGRAMS FOR FREQUENCY DISTRIBUTION 

```python
hs1 = grads['Sample_size'].hist()
hs1.set_ylabel('Frequency')
hs1.set_xlabel('Sample size')
hs1.set_title('Sample size frequency plot')
```

The majority of the sample size falls between 0 and 1000

```python
hs2 = grads['Median'].hist()
hs2.set_xlabel('Median')
hs2.set_ylabel('Frequency')
hs2.set_title('Median Frequency plot')
```

The majority of the median salary is high between 25000 and 45000 and highest between 25000 and 35000

```python
hs3 = grads['Employed'].hist()
hs3.set_ylabel('Frequency')
hs3.set_xlabel('Employed')
hs3.set_title('Employed frequency plot')
```

Majority of the employed majors falls between 0 and 50000

```python
hs4 = grads['Full_time'].hist()
hs4.set_ylabel("Frequency")
hs4.set_xlabel("Full time")
hs4.set_title("Full time frequency plot")
```

More Majors with full time job falls between 0 and 25000

```python
hs5 = grads['ShareWomen'].hist()
hs5.set_ylabel("Frequency")
hs5.set_xlabel('Sharewomen')
hs5.set_title('Sharewomen frequency plot')
```

```python
hs6 = grads['Unemployment_rate'].hist()
hs6.set_xlabel('Unemployment')
hs6.set_ylabel("Frequency")
hs6.set_title('Unemployment rate frequency plot')
```

The unemployment rate has the highest value between 0.05 and 0.075
